```python
from pyspark.sql.functions import *
policy_schema = "policy_id int,policy_type string,customer_id int,start_date timestamp,end_date timestamp,premium double,coverage_amount double"
df = spark.read.json("abfss://landing@policysysadls.dfs.core.windows.net/PolicyData", schema=policy_schema)
df_merge_flag = df.withColumn("merge_flag",lit(False))
df_merge_flag.write.option("path","abfss://bronzelayer@policysysadls.dfs.core.windows.net/Policy").mode("append").saveAsTable("bronzelayer.Policy")
```

```python
from datetime import datetime
def getFilePathWithDates(filePath):
    # get the current time in mm-dd-yyyy format
    current_time = datetime.now().strftime('%m-%d-%Y')
    new_file_path = filePath+'/'+current_time
    return new_file_path
```

```python
dbutils.fs.mv("abfss://landing@policysysadls.dfs.core.windows.net/PolicyData/",getFilePathWithDates("abfss://processed@policysysadls.dfs.core.windows.net/PolicyData/"),True)
```
