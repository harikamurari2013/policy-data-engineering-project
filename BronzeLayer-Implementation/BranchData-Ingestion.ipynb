```python
from pyspark.sql.functions import lit
branch_schema = "branch_id integer,branch_country string, branch_city string"
df = spark.read.parquet("abfss://landing@policysysadls.dfs.core.windows.net/BranchData/*.parquet",inferSchema=False,schema = branch_schema)
df_merge_flag = df.withColumn("maerge_flag",lit(False))
df_merge_flag.write.option("path","abfss://bronzelayer@policysysadls.dfs.core.windows.net/Branch").mode("append").saveAsTable("bronzelayer.Branch")
```

```python
from datetime import datetime
def getFilePathWithDates(filePath):
    # get the current time in mm-dd-yyyy format
    current_time = datetime.now().strftime('%m-%d-%Y')
    new_file_path = filePath+'/'+current_time
    return new_file_path
```
