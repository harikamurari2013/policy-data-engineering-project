```python
%sql
CREATE SCHEMA IF NOT EXISTS goldlayer LOCATION 'abfss://goldlayer@policysysadls.dfs.core.windows.net/';
```

###This table would contain the total sales for each policy type and each month. It would be used to analyze the performance of different policy types over time.

```python
%sql
create or replace table goldlayer.sales_by_policy_type_and_month(
  policy_type string,
  sale_month string,
  total_premium integer,
  updated_timestamp timestamp
) using delta location 'abfss://goldlayer@policysysadls.dfs.core.windows.net/sales_by_policy_type_and_month'
```

###This table would contain the number and amount of claims by policy type and claim status. It would be used to monitor the claims process and identify any trends or issues.

```python
%sql
create or replace table goldlayer.claims_by_policy_type_and_status(
  policy_type string,
  claim_status string,
  total_claims integer,
  total_claim_amount integer,
  updated_timestamp timestamp
) using delta location 'abfss://goldlayer@policysysadls.dfs.core.windows.net/claims_by_policy_type_and_status'
```

### Analyze the claim data based on the policy type like AVG, MAX, MIN, Count of claim.

```python
%sql
create or replace table goldlayer.claims_analysis(
  policy_type string,
  claim_status string,
  avg_claim_amount integer,
  max_claim_amount integer,
  min_claim_amount integer,
  total_claims integer,
  updated_timestamp timestamp
) using delta location 'abfss://goldlayer@policysysadls.dfs.core.windows.net/claims_analysis'
```
